```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.show()
```

```python
coloumns = ['Sepal length', 'Sepal width', 'Petal Length', 'Petal width', 'Class_labels']
df = pd.read_csv('iris.csv', names=coloumns, skiprows=1)
# Convert relevant columns to numeric
for col in coloumns[:-1]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.head(150)
```

```python
df.describe()
```

```python
sns.pairplot(df, hue='Class_labels')
```

```python
data = df.values

X = data[:,0:4]
Y = data[:,4]
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(X_test)
```

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
for i in range(len(prediction1)):
    print(prediction1[i], y_test[i])
```

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
prediction2 = model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train, y_train)
```

```python
prediction3 = model_DTC.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction3)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction2))
```

```python
X_new = np.array([[3, 2, 1, 0.2], [4.6, 2.3, 3.8, 2.2], [5.3, 2.7, 3.6, 1.9]])
prediction = model_svc.predict(X_new)
print("Prediction: {}".format(prediction))
```
